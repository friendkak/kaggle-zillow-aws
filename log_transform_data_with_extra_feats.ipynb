```python
import pandas as pd
df_properties_2016 =  pd.read_csv("s3://zillow-kaggle-awsml-2020/properties_2016.csv")
df_train_2016 = pd.read_csv('s3://zillow-kaggle-awsml-2020/train_2016_v2.csv')

df_properties_2017 =  pd.read_csv("s3://zillow-kaggle-awsml-2020/properties_2017.csv")
df_train_2017 = pd.read_csv('s3://zillow-kaggle-awsml-2020/train_2017.csv')

print("df_properties_2016: %s" % str(df_properties_2016.shape))
print("df_train_2016: %s" % str(df_train_2016.shape))
print("df_properties_2017: %s" % str(df_properties_2017.shape))
print("df_train_2017: %s" % str(df_train_2017.shape))
```

```python
merged_2016 = df_train_2016.merge(df_properties_2016, how='left', on='parcelid')
merged_2017 = df_train_2017.merge(df_properties_2017, how='left', on='parcelid')
```

```python
merged = merged_2016.append(merged_2017)

print(merged_2016.shape)
print(merged_2017.shape)
print(merged.shape)
```

```python
features = ['fireplacecnt', 'fullbathcnt',
            'garagecarcnt','calculatedbathnbr', 
            'bedroomcnt', 'bathroomcnt',
           'finishedsquarefeet12', 'garagetotalsqft',
           'yearbuilt']
log_transform_vars = ['bathroomcnt', 'bedroomcnt', 'bathroom_overage', 'bedroom_overage', 'calculatedfinishedsquarefeet', 
                      'square_foot_overage_small', 'square_foot_overage_large', 'square_foot_overage_opulent',
                     'basementsqft', 'bathroomcnt', 'bedroomcnt', 'calculatedbathnbr', 'threequarterbathnbr', 
                      'finishedfloor1squarefeet', 'calculatedfinishedsquarefeet', 'finishedsquarefeet6', 'finishedsquarefeet12', 
                      'finishedsquarefeet13', 'finishedsquarefeet15', 'finishedsquarefeet50', 'garagetotalsqft',
                      'garagecarcnt', 'lotsizesquarefeet', 'roomcnt', 'structuretaxvaluedollarcnt', 'landtaxvaluedollarcnt',
                      'taxamount', 'lot_size_overage_small', 'lot_size_overage_mid', 'lot_size_overage_large', 
                      'lot_size_overage_huge', 'lot_size_overage_raw'
                     ]

cat_vars = ['air_coded', 'heat_coded', 'fips' ]
#garage_size_level, fireplace_level,patio_level, basement_level, lotsizelevel

transaction_cols = ['parcelid','logerror', 'transactiondate']
```

```python
def code_data(data):
    rdata = pd.DataFrame({'air_coded':data['airconditioningtypeid'].map(lambda v: 'one' if v==1 else ('na' if pd.isna(v) else 'other'))})
    rdata['bathroom_overage'] = data['bathroomcnt'].map(lambda v: max(v-4,0))
    rdata['knockdown'] = data.apply(lambda row: 1 if row['landtaxvaluedollarcnt']>20000 and row['structuretaxvaluedollarcnt'] >20000 and row['structuretaxvaluedollarcnt'] < row['landtaxvaluedollarcnt'] else 0, axis='columns')
    rdata['unreliable_assessment'] = data.apply(lambda row: 1 if row['structuretaxvaluedollarcnt'] < 10000 and row['landtaxvaluedollarcnt'] > 10000 else 0, axis='columns')
    rdata['low_qual_building'] = data['buildingqualitytypeid'].map(lambda v: v in [2,3,5,6,8,9,11])
    rdata['med_qual_building'] = data['buildingqualitytypeid'].map(lambda v: v in [12])
    rdata['assessed_structure_cost_per_sqft_outlier'] = data.apply(lambda row: row['finishedsquarefeet12']>600 and row['structuretaxvaluedollarcnt']<2000000 and row['structuretaxvaluedollarcnt']/row['finishedsquarefeet12'] > 800, axis='columns')
    rdata['topmarket'] = data['taxvaluedollarcnt'].map(lambda v: 1 if v >= 5000000 else 0)
    rdata['bedroom_overage'] = data['bedroomcnt'].map(lambda v: max(v-6,0))
    rdata['lotsizelevel'] = data['lotsizesquarefeet'].map(lambda v: 0 if v<3000 or pd.isna(v) else (1 if v<5000 else (2 if v<8000 else (3 if v<12000 else (4 if v<20000 else (5 if v<43560 else (6 if v<100000 else (7 if v < 435600 else (8 if v<1000000 else 9)))))))))
    rdata['lowstructuretaxoutlier'] = data.apply(lambda row: not pd.isna(row['structuretaxvaluedollarcnt']) and not pd.isna(row['finishedsquarefeet12']) and row['structuretaxvaluedollarcnt']< row['finishedsquarefeet12']*200, axis='columns')
    rdata['lowlandtaxoutlier'] = data.apply(lambda row: not pd.isna(row['landtaxvaluedollarcnt']) and not pd.isna(row['lotsizesquarefeet']) and row['landtaxvaluedollarcnt'] < row['lotsizesquarefeet']*35, axis='columns')
    rdata['multifam'] = data['unitcnt'].map(lambda v: v>1)
    rdata['highdensity'] = data['unitcnt'].map(lambda v: v>4)
    rdata['typical_land_use'] = data['propertylandusetypeid'].map(lambda v: v in [261,266])
    rdata['single_story']=data['numberofstories']==1
    rdata['has_pool'] = data['poolcnt']>0
    rdata['istrunc_calculatedfinishedsquarefeet'] = data.calculatedfinishedsquarefeet.map(lambda x: x>8000)
    rdata['trunc_calculatedfinishedsquarefeet'] = data.calculatedfinishedsquarefeet.map(lambda x: min(x, 8000) if not pd.isna(x) else 0.0)
    rdata['istrunc_calculatedbathnbr'] = data.calculatedbathnbr.map(lambda x: x>5.0)
    rdata['trunc_calculatedbathnbr'] = data.calculatedbathnbr.map(lambda x: min(x,5.0) if not pd.isna(x) else 0.0)
    rdata['heat_coded'] = data['heatingorsystemtypeid'].map(lambda v: 'two' if v==2 else ('seven' if v==7 else 'other')) 
    rdata['has_pool'] = ~data.pooltypeid7.isna()
    rdata['no_garage_info'] = (data.garagecarcnt.isna() | data.garagecarcnt < 0) & (data.garagetotalsqft.isna() | data.garagetotalsqft < 0)
    rdata['garage_size_level'] = data.garagetotalsqft.map(lambda x: 0 if pd.isna(x) or x<=0 else (1 if x < 120 else (2 if x<250 else (3 if x<400 else 4 if x<600 else 5 if x<1000 else 6))))
    rdata['fireplace_level'] = data.fireplacecnt.map(lambda x: 0 if pd.isna(x) or x<=0 else (1 if x==1 else 2))
    rdata['patio_level'] = data.yardbuildingsqft17.map(lambda x: 0 if pd.isna(x) or x<=0 else (1 if x<80 else (2 if x<400 else 3)))
    rdata['has_large_accessory'] = data.yardbuildingsqft26.map(lambda x: x>=200)
    rdata['log_age'] = data.yearbuilt.map(lambda x: np.log(2020- min(x,2019)) if not pd.isna(x) else np.log(2020-1970))
    rdata['basement_level'] = data.basementsqft.map(lambda x: 0 if pd.isna(x) or x <=0 else( 1 if x<100 else (2 if x<500 else 3)))
    rdata['full_basement'] = data.apply(lambda row: ~pd.isna(row.basementsqft) and row.finishedfloor1squarefeet>100 and row.basementsqft/row.finishedfloor1squarefeet>.9, axis='columns')                                                    
    rdata['square_foot_overage_small'] = data.calculatedfinishedsquarefeet.map(lambda x: max(x-2000,0))
    rdata['square_foot_overage_large'] = data.calculatedfinishedsquarefeet.map(lambda x: max(x-4000,0))
    rdata['square_foot_overage_opulent'] = data.calculatedfinishedsquarefeet.map(lambda x: max(x-8000,0))
    rdata['recent_tax_delinquency'] = data.taxdelinquencyyear.map(lambda x: x>14 and x<=20)
    rdata['is_trunc_bedrooms'] = data.bedroomcnt.map(lambda x: x>6 if not pd.isna(x) else False)
    rdata['trunc_bedrooms'] = data.bedroomcnt.map(lambda x: min(x,6) if not pd.isna(x) else 0)
    rdata['is_trunc_bathrooms'] = data.bathroomcnt.map(lambda x: x>4 if not pd.isna(x) else False)
    rdata['trunc_bathrooms'] = data.bathroomcnt.map(lambda x: min(x,4) if not pd.isna(x) else 0)
    rdata['lot_size_overage_small'] = data.lotsizesquarefeet.map(lambda x: max(x-3000,0))
    rdata['lot_size_overage_mid'] = data.lotsizesquarefeet.map(lambda x: max(x-6000,0))
    rdata['lot_size_overage_large'] = data.lotsizesquarefeet.map(lambda x: max(x-10000,0))
    rdata['lot_size_overage_huge'] = data.lotsizesquarefeet.map(lambda x: max(x-20000,0)) 
    rdata['lot_size_overage_raw'] = data.lotsizesquarefeet.map(lambda x: max(x-40000,0))
    rdata['has_unreliable_bedrooms'] = data.bedroomcnt.isna() | data.bedroomcnt==0
    rdata['has_unreliable_bathrooms'] = data.bathroomcnt.isna() | data.bathroomcnt==0
    rdata['has_unreliable_calculatedfinishedsquarefeet'] = data.calculatedfinishedsquarefeet.isna() | data.calculatedfinishedsquarefeet==0
    return rdata
    
                                  
                              
```

```python
transformed_data_1 = pd.concat([merged[transaction_cols], code_data(merged), merged[set(log_transform_vars).intersection(merged.columns.values)]], axis='columns')
```

```python
transformed_data_1c = transformed_data_1.copy(deep=True)
transformed_data_1c[log_transform_vars] = transformed_data_1c[log_transform_vars].fillna(0.0).transform(lambda x: np.log(1.0+x.astype('float')))
```

```python
pandas_dt = pd.to_datetime(transformed_data['transactiondate'])
```

```python
cat_vars_in_transformed = set(cat_vars).intersection(transformed_data_1c.columns.values)
cat_vars_in_merged = ['fips']
transformed_data = pd.concat([transformed_data_1c.drop(cat_vars_in_transformed, axis='columns'), pd.get_dummies(transformed_data_1c[cat_vars_in_transformed], columns=cat_vars_in_transformed), pd.get_dummies(merged[cat_vars_in_merged], columns=cat_vars_in_merged), pd.get_dummies(pandas_dt.dt.to_period('M'))], axis='columns')
```
