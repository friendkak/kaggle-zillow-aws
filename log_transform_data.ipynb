We have 2,985,217 properties with 58 columns

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
import pandas as pd
df_properties_2016 =  pd.read_csv("s3://zillow-kaggle-awsml-2020/properties_2016.csv")
df_train_2016 = pd.read_csv('s3://zillow-kaggle-awsml-2020/train_2016_v2.csv')

df_properties_2017 =  pd.read_csv("s3://zillow-kaggle-awsml-2020/properties_2017.csv")
df_train_2017 = pd.read_csv('s3://zillow-kaggle-awsml-2020/train_2017.csv')

print("df_properties_2016: %s" % str(df_properties_2016.shape))
print("df_train_2016: %s" % str(df_train_2016.shape))
print("df_properties_2017: %s" % str(df_properties_2017.shape))
print("df_train_2017: %s" % str(df_train_2017.shape))
```

```python
df_properties = df_properties_2016.append(df_properties_2017)
df_train = df_train_2016.append(df_train_2017)

print(df_properties.shape)
print(df_train.shape)
```

```python
df = df_train.merge(df_properties, how='left', on="parcelid")
df.shape
```

```python
features = ['fireplacecnt', 'fullbathcnt',
            'garagecarcnt','calculatedbathnbr', 
            'bedroomcnt', 'bathroomcnt',
           'finishedsquarefeet12', 'garagetotalsqft',
           'yearbuilt']

transaction_cols = ['parcelid','logerror', 'transactiondate']
```

```python
si = SimpleImputer(fill_value=0, strategy='constant')
result = si.fit_transform(df[features].to_numpy())
```

```python
transaction_data = df[transaction_cols].to_numpy()
```

```python
transaction_data
```

```python
transaction_data.shape
```

```python
np.append(transaction_data, result, axis=1).shape
```

```python
df_result = pd.DataFrame(np.append(transaction_data, result, axis=1), columns=np.append(transaction_cols, features))
```

```python
df_result["homeage"] = df_result.apply(lambda x: int(2020 - x.yearbuilt), axis=1)
```

```python
updated_features = features[:-1]
updated_features.append("homeage")
updated_features
```

```python
df_result[updated_features] = df_result[updated_features].transform(lambda x: np.log(x.astype('float') + 1))
```

```python
df_result.to_csv("log_transformed_data.csv")
```
